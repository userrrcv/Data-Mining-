# BPN

## Objective: classify the tumor is benign or malign

### 1. Load Libraries

```python
# Import the breast cancer dataset from sklearn
from sklearn.datasets import load_breast_cancer

# Import function to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Import scaler to normalize the data (important for neural networks)
from sklearn.preprocessing import StandardScaler

# Import the neural network model (Backpropagation Neural Network)
from sklearn.neural_network import MLPClassifier

# Import decision tree model
from sklearn.tree import DecisionTreeClassifier

# Import evaluation metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Load the breast cancer dataset
data = load_breast_cancer()

# Extract input features (X)
X = data.data

# Extract target labels (y) -> 0 = malignant, 1 = benign
y = data.target


# Split the dataset into training (70%) and testing (30%)
# stratify=y keeps the same class distribution in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)


# Create a scaler object to normalize the features
scaler = StandardScaler()

# Fit the scaler on training data and transform it
X_train = scaler.fit_transform(X_train)

# Apply the same transformation to test data
X_test = scaler.transform(X_test)


# =========================
# BPN MODEL (Neural Network)
# =========================

# Create the neural network model
# hidden_layer_sizes=(10,) means 1 hidden layer with 10 neurons
# learning_rate_init controls how fast the model learns
# max_iter is the stopping condition (number of training iterations)
bpn = MLPClassifier(
    hidden_layer_sizes=(10,),
    learning_rate_init=0.01,
    max_iter=500,
    random_state=42
)

# Train the neural network using training data
bpn.fit(X_train, y_train)

# Make predictions on test data
y_pred_bpn = bpn.predict(X_test)

# Print accuracy of the neural network
print("BPN Accuracy:", accuracy_score(y_test, y_pred_bpn))

# Print detailed classification metrics (precision, recall, f1-score)
print("BPN Classification Report:")
print(classification_report(y_test, y_pred_bpn))

# Print confusion matrix to see prediction errors
print("BPN Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_bpn))
```
